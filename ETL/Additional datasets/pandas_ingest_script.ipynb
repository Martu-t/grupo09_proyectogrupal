```python
def new_features(day, lat, lon):
    '''
    Given as inputs the day of occurance, latitud and longitud of the event.
    fct returns temperature_2m_max, windspeed_10m_max and precipitation_sum at the corresponding place and day
    information gotten from https://open-meteo.com/en/docs API
    '''    
    query = 'https://archive-api.open-meteo.com/v1/era5?latitude='+lat+'&longitude='+lon+'&start_date='+day+'&end_date='+day+'&timezone=Greenwich&daily=temperature_2m_max&daily=windspeed_10m_max&daily=precipitation_sum'
    data = requests.get(query).json()
    tmax = data['daily']['temperature_2m_max'][0]
    windspeed = data['daily']['windspeed_10m_max'][0]
    precip = data['daily']['precipitation_sum'][0]

    return tmax, windspeed, precip


def ymd_to_day (year, mo, dy):
    '''
    Inputs three parameters and returns a date string in yyyy-mm-dd format
    '''

    y = str(year)
    if mo < 10:
        m = '0'+str(mo)
    else:
        m = str(mo)    
    if dy < 10:
        d = '0'+str(dy)
    else:
        d = str(dy)  
    day= y+'-'+m+'-'+d

    return day
```

```python
temp = []
wind = []
precip = []

end = len(earthquake_data)

for i in range(0,end):
    day = ymd_to_day(earthquake_data.loc[i,'Year'],earthquake_data.loc[i,'Mo'],earthquake_data.loc[i,'Dy'])
    lat = str(earthquake_data.loc[i,'Latitude'])
    lon = str(earthquake_data.loc[i,'Longitude'])
    t,w,p = new_features(day, lat, lon)
    temp.append(t)
    wind.append(w)
    precip.append(p)
    print( 'row: ',i,' OK! ---> ',round(((i+1)/end)*100,1),'% Done' )

# 11 min aprox
```

```python
earthquake_data['Temp_C'] = temp
earthquake_data['Wind_ms'] = wind
earthquake_data['Precip_mm'] = precip
```

```python
earthquake_data.to_csv('earthquakes_weather.csv', index=False)
```
